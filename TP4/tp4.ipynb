On peut déterminer le nombre de trames complet plus celle qui sont incomplete a la fin de la fenetre comme étant de :
$\frac{F_s + 1}{N_{hop}}$
Il faut après retirer les trames imcomplete a la fin donc on retire:
$\frac{N_{win}}{N_{hop}}$
Ce qui donne : $ \frac{F_s + 1 -N_{win}}{N_{hop}}$
